1. Reverse a NumPy array

```python
import numpy as np

arrA=np.array([1,2,3,6,4,5])
revArr=arrA[::-1]
print(revArr)
```

2. Compare two NumPy arrays

```python
import numpy as np

matA=np.array([[1,2],[3,4]])
matB=np.array([[1,2],[3,4]])

isEqual=np.array_equal(matA,matB)
print(isEqual)
```

3. Find the most frequent value and its indices
(i) Array x

```python
import numpy as np

xArr=np.array([1,2,3,4,5,1,2,1,1,1])

vals,counts=np.unique(xArr,return_counts=True)
maxVal=vals[np.argmax(counts)]
indices=np.where(xArr==maxVal)

print("Most Frequent Value:",maxVal)
print("Indices:",indices)
```

(ii) Array y

```python
import numpy as np

yArr=np.array([1,1,1,2,3,4,2,4,3,3])

valsY,countsY=np.unique(yArr,return_counts=True)
freqVal=valsY[np.argmax(countsY)]
idx=np.where(yArr==freqVal)

print("Most Frequent Value:",freqVal)
print("Indices:",idx)
```

4. Matrix operations

```python
import numpy as np

mat=np.matrix([[4,1,9],[12,3,1],[4,5,6]])

totalSum=np.sum(mat)
rowSum=np.sum(mat,axis=1)
colSum=np.sum(mat,axis=0)

print("Sum of all elements:",totalSum)
print("Row-wise sum:",rowSum)
print("Column-wise sum:",colSum)
```
